```sql
USE CATALOG banking_catalog;
USE SCHEMA bronze;

-- Customers
CREATE OR REPLACE TABLE customers AS
SELECT *
FROM read_files(
  '/Volumes/banking_catalog/bronze/raw_ingestion_volume/customers.csv',
  format => 'csv',
  header => true,
  inferSchema => true
);

-- Accounts
CREATE OR REPLACE TABLE accounts AS
SELECT *
FROM read_files(
  '/Volumes/banking_catalog/bronze/raw_ingestion_volume/accounts.csv',
  format => 'csv',
  header => true,
  inferSchema => true
);

-- Transactions
CREATE OR REPLACE TABLE transactions AS
SELECT *
FROM read_files(
  '/Volumes/banking_catalog/bronze/raw_ingestion_volume/transactions.csv',
  format => 'csv',
  header => true,
  inferSchema => true
);
```
